```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def kmeansImagen(imgOriginal, nClusters):
    img = cv2.imread(imgOriginal) #Leer imagen

    vectorImg = img.reshape(-1, 3) #Vectorizar, RGB

    kmeans = KMeans(n_clusters=nClusters) #KMeans
    kmeans.fit(vectorImg)

    labels = kmeans.predict(vectorImg) #Asignar colores a los centros
    vectorImg2 = kmeans.cluster_centers_[labels]

    img2 = vectorImg2.reshape(img.shape)  #Desvectorizar

    return img2.astype(np.uint8)

imgOriginal = "si.jpg" 

nClusters = [2, 4, 6, 8, 12, 18, 32, 64]

plt.figure(figsize=(15, 10))

plt.subplot(3, 3, 1)
plt.imshow(cv2.cvtColor(cv2.imread(imgOriginal), cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Imagen original')

for i, n in enumerate(nClusters, start=2):
    imgNueva = kmeansImagen(imgOriginal, n)
    plt.subplot(3, 3, i)
    plt.imshow(cv2.cvtColor(imgNueva, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f'Imagen nueva {i-1} con {n} clusters')

plt.tight_layout()
plt.show()
```
